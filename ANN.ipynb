```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
```

```python
data = make_moons()
print(data)
X = data[0]
Y = data[1]
```

```python
Y = Y.reshape(-1,1)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
def binary_cross_entropy(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1-y_true)*np.log(1-y_pred))
```

```python
input_dim = 2
hidden_dim = 4
output_dim = 1
learning_rate = 0.1

def initialise_parameters():
    np.random.seed(42)
    weights_ih = np.random.randn(input_dim, hidden_dim)
    bias_h = np.zeros((1,hidden_dim))
    weights_ho = np.random.randn(hidden_dim, output_dim)
    bias_o = np.zeros((1, output_dim))
    print(weights_ih, bias_h, weights_ho, bias_o)

    return weights_ih, bias_h, weights_ho, bias_o
    
```

```python
def forward_propagation(X, weights_ih, bias_h, weights_ho, bias_o):
    hidden_input = np.dot(X, weights_ih) + bias_h
    hidden_output = sigmoid(hidden_input)
    output_input = np.dot(hidden_output, weights_ho) + bias_o
    output = sigmoid(output_input)

    return hidden_output, output
```

```python
def backward_propagation(X, Y, hidden_output, output, weights_ho):
    m = X.shape[0]
    output_error = output - Y
    output_gradient = output_error * sigmoid_derivative(output)

    hidden_error = np.dot(output_gradient, weights_ho.T)
    hidden_gradient = hidden_error * sigmoid_derivative(hidden_output)

    d_weights_ih = np.dot(X.T, hidden_gradient) / m
    d_bias_h = np.sum(hidden_gradient, axis=0, keepdims=True) / m
    d_weights_ho = np.dot(hidden_output.T, output_gradient) / m
    d_bias_o = np.sum(output_gradient, axis = 0, keepdims = True) / m

    return d_weights_ih, d_bias_h, d_weights_ho, d_bias_o
```

```python
weights_ih, bias_h, weights_ho, bias_o = initialise_parameters()
```

```python
for epoch in range(100):
    hidden_output, output = forward_propagation(X_train, weights_ih, bias_h, weights_ho, bias_o)
    loss = binary_cross_entropy(Y_train, output)

    d_weights_ih, d_bias_h, d_weights_ho, d_bias_o = backward_propagation(X_train, Y_train, hidden_output, output, weights_ho)
    weights_ih -= d_weights_ih * learning_rate
    bias_h -= d_bias_h * learning_rate
    weights_ho -= d_weights_ho * learning_rate
    bias_o -= d_bias_o * learning_rate

    if epoch%10 == 0:
        print(f"Epoch: {epoch} Loss = {loss:.4f}")
```

```python
hidden_output, output = forward_propagation(X_test, weights_ih, bias_h, weights_ho, bias_o)
final_loss = binary_cross_entropy(Y_test, output)
print(f"Final Loss = {final_loss:.4f}")
```
