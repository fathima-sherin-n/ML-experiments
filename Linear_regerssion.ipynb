```python
maths_score_mean = np.mean(X_train)
maths_score_median = np.median(X_train)
maths_score_min = np.min(X_train)
maths_score_max = np.max(X_train)

print("Mean",maths_score_mean)
print("Median",maths_score_median)
print("Min",maths_score_min)
print("Max",maths_score_max)
```

```python
def fit(X, Y, learning_rate, iterations):
    no_samples, no_features = X.shape
    W = np.zeros(no_features)
    b = 0
    print("W", W, b)
    
    for _ in range(iterations):
#         print("X", X)
#         print("Y", Y)
        W,b = update_weights(X, Y, W, b, learning_rate, no_samples)
    return W, b
```

```python
def update_weights(X, Y, W, b, learning_rate, no_samples):
#     print("X shape", X.shape)
#     print("Y shape", Y.shape)
#     print("W shape", W.shape)
    Y_pred = np.dot(X,W) + b
    print("Y_pred - Y", Y_pred - Y)
    dW = (-2/no_samples) * (np.dot(X.T, Y_pred - Y))
    db = (2/no_samples) * (np.sum(Y_pred - Y))
    print("dW", dW)
    W -= learning_rate * dW
    b -= learning_rate * db
    return W, b
```

```python
W, b = fit(X_train, Y_train, 0.001, 100)
print(W, b)
```

```python
def lin_pred(X, W, b):
    return np.dot(W,X) + b
```

```python
pred_s1 = lin_pred(50, W, b)
print("Predicted writing score for 50", pred_s1[0])
```

```python
pred_s1 = lin_pred(75, W, b)
print("Predicted writing score for 75", pred_s1[0])
```
